### Pattern Observations

```python
def extract_feature_patterns(df, target_col="label"):
    observations = []

    # Correlations Numeric Variables
    numeric_df = df.select_dtypes(include=[np.number])
    if target_col in numeric_df.columns:
        corr = numeric_df.corr()[target_col].sort_values(ascending=False)
        top_corr = corr.drop(target_col).head(3)
        observations.append(f"Strong correlations detected with target: {dict(top_corr.round(3))}")

    # Correlation between key continuous variables
    if {"frp", "brightness", "bright_t31"}.issubset(df.columns):
        corr_frp = df[["frp","brightness","bright_t31"]].corr().round(3)
        observations.append(f"FRP, Brightness, and Bright_T31 show strong mutual correlations:\n{corr_frp.to_dict()}")

    # Spatial clustering
    if {"latitude", "longitude"}.issubset(df.columns):
        erupt_locs = df[df[target_col]==1][["latitude","longitude"]]
        spread_lat = erupt_locs["latitude"].std()
        spread_lon = erupt_locs["longitude"].std()
        observations.append(
            f"Eruption events cluster spatially near mean(lat={erupt_locs['latitude'].mean():.2f}, "
            f"lon={erupt_locs['longitude'].mean():.2f}) with std dev (lat={spread_lat:.2f}, lon={spread_lon:.2f})."
        )

    # Temporal spikes
    if "year" in df.columns:
        erupt_years = df[df[target_col]==1]["year"].value_counts().sort_index()
        if not erupt_years.empty:
            peak_year = erupt_years.idxmax()
            observations.append(f"Eruption frequency peaks in year {peak_year} with {erupt_years.max()} events.")

    if "month" in df.columns:
        erupt_months = df[df[target_col]==1]["month"].value_counts().sort_index()
        if not erupt_months.empty:
            peak_month = erupt_months.idxmax()
            observations.append(f"Eruption activity shows seasonal influence, peaking in month {peak_month}.")

    # Confidence distribution
    if "confidence" in df.columns:
        mean_conf = df["confidence"].mean()
        observations.append(f"Average confidence of detections is {mean_conf:.2f}, "
                            f"indicating generally {'high' if mean_conf>50 else 'low'} reliability.")

    print("PATTERN Observations:")
    for obs in observations:
        print(f"- {obs}")

extract_feature_patterns(df, target_col="label")
```

### Feature Engineering & Preprocessing

```python
X = df.drop("label", axis=1)
y = df["label"]
```

```python
# Compute Correlations
numeric_cols = X.select_dtypes(include=["float64", "int64"]).columns.tolist()
corr_matrix = df[numeric_cols + ["label"]].corr()

corr_with_target = corr_matrix["label"].drop("label").abs().sort_values(ascending=False)
print("Correlation of Features with Target:\n")
print(corr_with_target)
```

```python
# Features Selection
top_features = corr_with_target.head(8).index.tolist()
X_selected = X[top_features]

print("\nSelected Features based on correlation:", top_features)
```

### Train-Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_selected, y, test_size=0.2, random_state=42, stratify=y
)
```

### Data Normalization using MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Data Balancing with SMOTE

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train_scaled, y_train)

print("\nBefore SMOTE:", y_train.value_counts().to_dict())
print("After SMOTE:", y_train_res.value_counts().to_dict())
```

### Build CNN Model

```python
import tensorflow as tf
from tensorflow.keras import layers, models

cnn_model = models.Sequential([
    layers.Input(shape=(X_train_res.shape[1], 1)),
    layers.Conv1D(32, kernel_size=2, activation="relu"),
    layers.MaxPooling1D(pool_size=2),
    layers.Flatten(),
    layers.Dense(64, activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(1, activation="sigmoid")
])

cnn_model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])

# Print model summary
print("\nCNN Model Summary:\n")
cnn_model.summary()
```

### Training CNN (with Early Stopping)

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(monitor="val_loss", patience=5, restore_best_weights=True)

history = cnn_model.fit(
    X_train_res.reshape(-1, X_train_res.shape[1], 1),
    y_train_res,
    validation_data=(X_test_scaled.reshape(-1, X_test_scaled.shape[1], 1), y_test),
    epochs=50,
    batch_size=32,
    callbacks=[early_stop],
    verbose=1
)
```

### Train RandomForest Model

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=200, max_depth=8, random_state=42
)

rf_model.fit(X_train_res, y_train_res)
```

### Ensembling of the Models

```python
cnn_preds = cnn_model.predict(X_test_scaled.reshape(-1, X_test_scaled.shape[1], 1)).ravel()

rf_preds = rf_model.predict_proba(X_test_scaled)[:,1]

ensemble_preds = (cnn_preds + rf_preds) / 2
ensemble_classes = (ensemble_preds > 0.5).astype(int)
```

### Evaluation of Ensemble Model

```python
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score

acc = accuracy_score(y_test, ensemble_classes)
roc_auc = roc_auc_score(y_test, ensemble_preds)

# Classification Report
print("\nClassification Report:\n", classification_report(y_test, ensemble_classes, target_names=["No Eruption", "Eruption"]))
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, ensemble_classes, labels=[0, 1])

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["No Eruption", "Eruption"])
disp.plot(cmap="Blues", values_format="d")
plt.title("Confusion Matrix")
plt.show()
```

```python
history_dict = history.history

# Plot Accuracy
plt.figure(figsize=(6,5))
plt.plot(history_dict['accuracy'], label='Train Accuracy', linewidth=2)
plt.plot(history_dict['val_accuracy'], label='Validation Accuracy', linewidth=2)
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Training & Validation Accuracy")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()

# Plot Loss
plt.figure(figsize=(6,5))
plt.plot(history_dict['loss'], label='Train Loss', linewidth=2)
plt.plot(history_dict['val_loss'], label='Validation Loss', linewidth=2)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training & Validation Loss")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

```python
# ROC Curve
from sklearn.metrics import roc_curve, auc

fpr, tpr, _ = roc_curve(y_test, ensemble_preds)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, label=f"Ensemble AUC = {roc_auc:.3f}", linewidth=2)
plt.plot([0,1],[0,1],"--", color="gray")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - CNN + RandomForest Ensemble")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```

```python
import shap

def ensemble_predict(X):
    # CNN prediction
    cnn_p = cnn_model.predict(X.reshape(-1, X.shape[1], 1)).ravel()
    rf_p = rf_model.predict_proba(X)[:, 1]
    return (cnn_p + rf_p) / 2

background = shap.sample(X_train_scaled, 100, random_state=42)
explainer = shap.KernelExplainer(ensemble_predict, background)

# ----------------------------
#  Compute SHAP values
X_sample = shap.sample(X_test_scaled, 200, random_state=42)

shap_values = explainer.shap_values(X_sample)
```

```python
shap.summary_plot(shap_values, X_sample, feature_names=top_features, plot_size=(8,5))
```

```python
mean_abs_shap = np.mean(np.abs(shap_values), axis=0)

if len(mean_abs_shap) == len(top_features):
    importance = pd.Series(mean_abs_shap, index=top_features)
else:
    importance = pd.Series(mean_abs_shap, index=[f"f{i}" for i in range(len(mean_abs_shap))])

importance = importance.sort_values(ascending=True)

plt.figure(figsize=(7,5))
importance.plot(kind="barh", color="royalblue", edgecolor="black")
plt.title("Global Feature Importance (Ensemble SHAP)")
plt.xlabel("Mean |SHAP Value|")
plt.tight_layout()
plt.show()
```
